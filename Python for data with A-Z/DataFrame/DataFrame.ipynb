# Introduction to Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"]= 8,4
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
stat.head()
```

```python
# Distribution
vis1 = sns.distplot(stat["Total"])
```

```python
vis1 = sns.distplot(stat["Total"],bins=10)
```

```python
# Boxplots:
vis2 = sns.boxplot(data = stat, x = "Gold",y = "Total")
```

```python
# Seaborn Gallery to read on the documentation for better understad
```

### Visualizing with Seaborn

```python
vis3 = sns.lmplot(data = stat,x = "Rank",y = "Total",fit_reg=False)
```

```python
vis3 = sns.lmplot(data = stat,x = "Rank",y = "Total",fit_reg=False,hue= None,size = 10)
```

### Keyword Argument with python (advanced tutorial)

```python
# Marke size:
vis3 = sns.lmplot(data = stat,x = "Rank",y = "Total",fit_reg=False,hue= None,size = 8,
                 scatter_kws={"s":100})
```
